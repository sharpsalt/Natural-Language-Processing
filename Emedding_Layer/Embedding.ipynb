```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
##sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]#These are my set of questions which i am going to specially use
```

```python
#Define the vocabulary size
voc_size=10000
```

```python
##One Hot Representation for every word
onehot_repr=[one_hot(words,voc_size)for words in sent]
print(onehot_repr)
```

```python
#Word Embedding Representation

from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
#Pad_seqeunce is very important
#since we will build sequential model so we also import sequential
```

```python
import numpy as np
```

```python
#Pad Seqeunce
#___________
#Observe each and every sentence like not every sentence have same number of words so, we need to make all these sentence of equal size else we will not be able to train in our RNN, because at the end of day for training in RNN it will go with th fixed number for words for a certain fixed number of timestamps
```

```python
sent_length=8 #becase i know all the words have an average size of 5
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
#Feature Representation
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.build(input_shape=(None, sent_length))
model.summary()
```

```python
model.predict(embedded_docs)
```
